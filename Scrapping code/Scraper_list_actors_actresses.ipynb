```python
donnees=[]
def get_all_pages():
    #on cre une liste pour mettre les urls
    urls=[]
    page_number=51
    
    for i in range(200):
        i=f"https://www.imdb.com/search/name/?match_all=true&start={page_number}&ref_=rlm"
        page_number +=50
        urls.append(i)
        #print(i)
    return urls

#Function to scrape the name of the actresseses and actors
def parse_name(url):
    r= requests.get(url)
    soup1=BeautifulSoup(r.content,"html.parser")

    #print(soup1)
    container_html=soup1.findAll('div',class_="lister-item mode-detail")
    #print(names)
    
    
    for enter_container in container_html:
        
        #Gestion des erreurs au cas ou il y a aucun nom dans la balise
        try:
            names=enter_container.find('h3',class_="lister-item-header").find('a').text.strip()
        except AttributeError as e:
            names=""
        
        """try:
            links+=enter_container.find('h3',class_="lister-item-header").find('a')['href']
        except AttributeError as e:
            links+=""
        """  
        #profession=enter_container.find('div',class_="lister-item-content").find('p',class_="text-muted text-small").text
        
        try:
            film_shot=enter_container.find('div',class_="lister-item-content").find('p',class_="text-muted text-small").find('a').text.strip()
        except AttributeError as e:
            film_shot=""
        
        try:
            photo=enter_container.find('div',class_="lister-item-image").find('a').find('img')['src']
        except AttributError as e:
            photo=""
        
        donnees.append({'name':names,'films':film_shot,'Photo':photo})
       
        
    #print(data[0:2])
            
def parse_list_scrape():
    #on recupere les urls pour faire une boucle for pour passer sur chacune des urls les unes a la suite des autres
    pages=get_all_pages()
    
    for page in pages:
        parse_name(url=page)
        print(f"on scrape :  {page}")#indiquer la page dans laquelle nous sommes en train de scraper
    print(donnees)   
   
```
